## Bisection Search Algorithm:

```python
print('Please think of a number between 0 and 100 ')
low = 0
high = 100
medium = (high+low)//2
state = True
while state:
    print("Is your secret number is " + str(medium))
    guess = input("Enter 'h' to indicate the gess is to high..Enter 'l' to indicate the gess is to low..Enter 'c' to indicate the gess is to c")
    if guess == 'c':
        print("Game Over.. Your secret number was :"+str(medium))
        state = False
    elif guess =='h':
        high = medium
        medium = (high+low)//2
    elif guess == 'l':
        low = medium
        medium = (high+low)//2
    else:
        print("Sorry, i don'y understand you")
```
